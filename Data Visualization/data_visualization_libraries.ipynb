<h1><u>DATA VISUALIZATION LIBRARIES</u></h1>
<p>This notebook highlights the various libraries such as <b>seaborn</b> and <b>matplotlib</b> used in Machine learning for data visualization</p>

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.plotting.register_matplotlib_converters()
import seaborn as sns
%matplotlib inline
data_iris_path= "S:\playing with data with python\kaggle course for _intro to machine learning_ and other resourses\input_data\iris.csv"
iris_data = pd.read_csv(data_iris_path,index_col="Species")
iris_data.describe()
```

<h3>LINE PLOT</h3>

```python
plt.figure(figsize=(16,6))          # This creates the size of the graph canvas
plt.title("IRIS LINE PLOT")         # This sets the title of the plot
sns.lineplot(data=iris_data["SepalLengthCm"])        # This plots the plot
sns.lineplot(data=iris_data["SepalWidthCm"])
sns.lineplot(data=iris_data["PetalLengthCm"])
sns.lineplot(data=iris_data["PetalWidthCm"])
plt.xlabel("Species: 0 , 1 , 2")
```

<p>Here we print the <b>line plot </b>of all the sepal and petal details of various species</p>

<h3>BAR PLOT</h3>

```python
plt.figure(figsize=(15,5))
sns.barplot(data=iris_data)
```

<h3>HEAT MAP</h3>

```python
plt.figure(figsize=(15,5))
sns.heatmap(data=iris_data)
```

<h2>HISTOGRAMS & DENSITY PLOTS</h2>
<p>Next we will be dealing with the histograms and density plots
</p>

<p>To create a <b>Histogram</b> use the sns.distplot() method</p>

```python
sns.distplot(a=iris_data["SepalWidthCm"],kde=False)
```

<p><b>Kde = False</b></p>kde is set to false for a histogram. If it is set to true then we will get a smoothed histogram

<b>Kde = True</b>

```python
sns.distplot(a=iris_data["SepalWidthCm"],kde=True)
```

<h3>DENSITY PLOT</h3>
<p>We will be creating the kernel density plot(kde)</p>

```python
sns.kdeplot(data=iris_data["PetalLengthCm"],shade=True)
```

<h3>2D KDE PLOTS</h3>
<p>We are not restricted to single variable plots and we can even plot 2 variable density plots</p>

```python
sns.jointplot(x=iris_data["PetalLengthCm"],y=iris_data["SepalWidthCm"],kind="kde")
```

<p>This plot looks very nice and provides insight into strong relation between two variables</p>

<h3>COLOUR CODED PLOTS</h3>

```python
sns.distplot(a=iris_data["SepalWidthCm"],kde=False)
sns.distplot(a=iris_data["PetalWidthCm"],kde=False)
sns.distplot(a=iris_data["SepalLengthCm"],kde=False)
sns.distplot(a=iris_data["PetalLengthCm"],kde=False)
```

```python
sns.kdeplot(data=iris_data["PetalLengthCm"],shade=True)
sns.kdeplot(data=iris_data["SepalLengthCm"],shade=True)
sns.kdeplot(data=iris_data["PetalWidthCm"],shade=True)
sns.kdeplot(data=iris_data["SepalWidthCm"],shade=True)
```

<p>This type of plots help us to separate the features that are very much useful in the model selection</p>

<ul>
<li><b>Trends</b> - A trend is defined as a pattern of change.
<br><b>sns.lineplot</b> - Line charts are best to show trends over a period of time, and multiple lines can be used to show trends in more than one group.</li>
<br>
<li><b>Relationship</b> - There are many different chart types that you can use to understand relationships between variables in your data.
<br><b>sns.barplot</b> - Bar charts are useful for comparing quantities corresponding to different groups.
<br><b>sns.heatmap </b>- Heatmaps can be used to find color-coded patterns in tables of numbers.
<br><b>sns.scatterplot</b> - Scatter plots show the relationship between two continuous variables; if color-coded, we can also show the relationship with a third categorical variable.
<br><b>sns.regplot</b> - Including a regression line in the scatter plot makes it easier to see any linear relationship between two variables.
<br><b>sns.lmplot </b>- This command is useful for drawing multiple regression lines, if the scatter plot contains multiple, color-coded groups.
<br><b>sns.swarmplot</b> - Categorical scatter plots show the relationship between a continuous variable and a categorical variable.
</li>
<br>
<li><b>Distribution</b> - We visualize distributions to show the possible values that we can expect to see in a variable, along with how likely they are.
<br><b>sns.distplot</b> - Histograms show the distribution of a single numerical variable.
<br><b>sns.kdeplot</b> - KDE plots (or 2D KDE plots) show an estimated, smooth distribution of a single numerical variable (or two numerical variables).
<br><b>sns.jointplot</b> - This command is useful for simultaneously displaying a 2D KDE plot with the corresponding KDE plots for each individual variable.
</li>
</ul>
